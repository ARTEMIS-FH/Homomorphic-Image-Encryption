```python
import Paillier
import ImageCryptography
```

```python
public_key = Paillier.PublicKey(3599)
private_key = Paillier.PrivateKey(59,61,3599)
```

```python
img = ImageCryptography.Image.open("test-images/College_ID.jpeg")
img.show()
```

```python
cimg = ImageCryptography.ImgEncrypt(public_key,img)
```

```python
print(cimg)
```

```python
ImageCryptography.saveEncryptedImg(cimg,'test.bin')
```

```python
b_img = ImageCryptography.homomorphicBrightness(public_key,cimg,200)
img_b = ImageCryptography.ImgDecrypt(public_key,private_key,b_img)
img_b
```

```python

dimg = ImageCryptography.loadEncryptedImg('test.bin')
img_d = ImageCryptography.ImgDecrypt(public_key, private_key, dimg)
```
